# Задача №1

### Часть 1

Построить в общих осях графики для:

- $\alpha = 1, \beta = 1$

- $\alpha = 2, \beta = 1$

- $\alpha = 1, \beta = 2$

```python
x = np.linspace(-5, 5, 61)
def f(x, alfa=1, beta=1):
    with np.errstate(divide='ignore', invalid='ignore'):
        return (x**beta + alfa**beta)/x**beta
def f0(x):
    return x*0

plt.figure(figsize=(13,8))
plt.plot(x, f(x), 'r', 
         x, f(x, alfa=2), 'b', 
         x, f(x, beta=2), 'g')
plt.xlabel('x', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.title(u'График функции $f(x) = \\frac{x^{\\beta}+\\alpha^{\\beta}}{x^{\\beta}}$ ' + \
            'при различных $\\alpha$ и $\\beta$', fontsize=18)
plt.legend(["$\\alpha = 1, \\beta = 1$", 
            "$\\alpha = 2, \\beta = 1$", 
            "$\\alpha = 1, \\beta = 2$"])
plt.savefig('Задача1_1.svg')
plt.show()
```

### Часть 2

Построить в общих осях графики для $x>0$.

На том же графике сделать 2 врезки, демонстрирующие поведение графиков на 2 интервалах:

- для малых x

- для больших x

Необходимо продемонстрировать возможность (или невозможность) пересечений и стремление функций.

Цвет линий на врезках и основном графике должен быть одинаковым для одних и тех же значений $\alpha$ и $\beta$.

```python
x = np.arange(0, 10, 0.25)
x_min = np.arange(0, 2, 0.25)
x_max = np.arange(20, 1000, 1)
fig = plt.figure(figsize=(13,8), dpi=300)


axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(x, f(x), 'r', 
         x, f(x, alfa=2), 'b', 
         x, f(x, beta=2), 'g')
plt.xlabel('x', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.legend(["$\\alpha = 1, \\beta = 1$", 
            "$\\alpha = 2, \\beta = 1$", 
            "$\\alpha = 1, \\beta = 2$"])
plt.title(u'График функции $f(x) = \\frac{x^{\\beta}+\\alpha^{\\beta}}{x^{\\beta}}$ ' + \
            'при различных $\\alpha$ и $\\beta$ ' + \
            'при $x > 0$', fontsize=18)


axes2 = fig.add_axes([0.15, 0.4, 0.35, 0.35])
axes2.plot(x_min, f(x_min), 'r', 
         x_min, f(x_min, alfa=2), 'b', 
         x_min, f(x_min, beta=2), 'g')
plt.xlabel('x', fontsize=10)
plt.ylabel('f(x)', fontsize=10)
axes2.grid(which='major',
        color = 'k')
axes2.minorticks_on()
axes2.grid(which='minor',
        color = 'gray',
        linestyle = ':')
plt.title(u'Графики функции $f(x) = \\frac{x^{\\beta}+\\alpha^{\\beta}}{x^{\\beta}}$ ' + 
            'для малых $x$', fontsize=12)


axes3 = fig.add_axes([0.6, 0.4, 0.35, 0.35])
axes3.plot(x_max, f(x_max), 'r', 
         x_max, f(x_max, alfa=2), 'b', 
         x_max, f(x_max, beta=2), 'g')
plt.xlabel('x', fontsize=10)
plt.ylabel('f(x)', fontsize=10)
plt.title(u'Графики функции $f(x) = \\frac{x^{\\beta}+\\alpha^{\\beta}}{x^{\\beta}}$ ' + 
            'для больших $x$', fontsize=12)
plt.savefig('Задача1_2.svg', bbox_inches = "tight");
```

### Выводы:
Видно, что для $x=1$ графики функции при $\alpha=1$ и $\beta=1$ или $2$ пересекаются в точке $f(x)=2$.

Также можно отметить, что в точке $f(x)=5$ графики пересекаются при $\beta=1$ и $\alpha=1$ и $2$ для $x=0.5$.

Для больших $x$ видно, что функция стремится к значению $f(x)=1$.

### Часть 3

Построить в общих осях графики для $x<0$.

На том же графике сделать 1 врезку, демонстрирующую поведение графиков при удалении x от 0 к $-\infty$.

Необходимо продемонстрировать возможность (или невозможность) пересечений и стремление функций. Так же нанесите на графики прямую f(x) = 0.

Цвет линий на врезках и основном графике должен быть одинаковым для одних и тех же значений $\alpha$ и $\beta$.

```python
x = np.arange(-5, 0, 0.25)
x_min = np.linspace(-1000, 0, 100)
fig = plt.figure(figsize=(13,8), dpi=300)

axes_main = fig.add_axes([0, 0, 1, 1])
axes_main.plot(x, f(x), 'r',
           x, f(x, alfa=2), 'b',
           x, f(x, beta=2), 'g', 
           x, f0(x), 'k')
plt.legend(["$\\alpha = 1, \\beta = 1$", 
            "$\\alpha = 2, \\beta = 1$", 
            "$\\alpha = 1, \\beta = 2$",
            "$f(x)=0$"])
plt.xlabel('x', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.title(u'График функции $f(x) = \\frac{x^{\\beta}+\\alpha^{\\beta}}{x^{\\beta}}$ ' + \
            'при различных $\\alpha$ и $\\beta$ ' + \
            'при $x < 0$', fontsize=18);


axes_minor = fig.add_axes([0.3, 0.5, 0.4, 0.4])
axes_minor.plot(x_min, f(x_min), 'r', 
         x_min, f(x_min, alfa=2), 'b', 
         x_min, f(x_min, beta=2), 'g',
         x_min, f0(x_min), 'k')
plt.legend(["$\\alpha = 1, \\beta = 1$", 
            "$\\alpha = 2, \\beta = 1$", 
            "$\\alpha = 1, \\beta = 2$",
            "$f(x)=0$"])
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.title(u'Графики функции $f(x) = \\frac{x^{\\beta}+\\alpha^{\\beta}}{x^{\\beta}}$ ' + \
            'при отрицательных $x$', fontsize=12)
plt.savefig('Задача1_3.svg', bbox_inches = "tight");
```

Сделайте выводы о поведении графиков при различных значениях $\alpha$ и $\beta$.

Изучите возможности пересечения и то к чему стремятся кривые.

Сделайте выводы о возрастании/убывании.

### Выводы:
При $\beta=2$ для малых значений $x$ поведение функции закономерно - она резко возрастает, так как нахождение в знаменателе малого числа, еще и возведенного в квадрат увеличивает значение функции в разы, аналогично функции $f(x)=\frac{1}{x^2}$

Также видны точки пересечения с $f(x)=0$ для $x=-1$ и $-2$ при $\beta=1$ и $\alpha=1$ и $2$

Для $x<0$ функция стремится к значению $f(x)=1$.

- $\alpha = 0, \beta = 2$

## Задача №3

Построить в общих осях графики для:

- $\alpha = 1, \beta = 0.5$

- $\alpha = 1, \beta = -0.5$

- $\alpha = 1, \beta = -1.5$

Сделайте выводы о поведении графиков, включая возрастание/убывание и выпуклость/вогнутость

```python
x3 = np.arange(-10, 10, 0.1)
fig = plt.figure(figsize=(13,8), dpi=300)
axes3 = fig.add_axes([0, 0, 1, 1])
axes3.plot(x3, f(x3, beta=0.5), 'c',
          x3, f(x3, beta=-0.5), 'm',
          x3, f(x3, beta=-1.5), 'b')
axes3.grid(which='major',
        color = 'k')
axes3.minorticks_on()
axes3.grid(which='minor',
        color = 'gray',
        linestyle = ':')
plt.title(u'График функции $f(x) = \\frac{x^{\\beta}+\\alpha^{\\beta}}{x^{\\beta}}$ ' +
            'при различных $\\alpha$ и $\\beta$', fontsize=16)
axes3.legend(["$\\alpha=1, \\beta=0.5$", "$\\alpha=1, \\beta=-0.5$", 
             "$\\alpha=1, \\beta=-1.5$"], fontsize=12)
plt.xlabel('x', fontsize=14)
plt.ylabel('f(x)', fontsize=14)

plt.savefig('Задача3.svg', bbox_inches = "tight");
```

Для $\beta=0.5$ функция ведет себя как и для $\beta>1$ $-$ возрастает при $x$ стремящемся к нулю. Для больших $x$ она стремится к $1$.

Для $\beta=-0.5$ функция монотонно возрастает, является вогнутой (выпуклой вверх), для малых $x$ стремится к $1$.

Для $\beta=-1.5$ функция возрастает заметно быстрее, является выпуклой (выпуклой вниз), для малых $x$ стремится к $1$.

Общей точкой для всех графиков является $f(x)=2$ при $x=1$, что неудивительно, так как единица в любой степени остается единицей, а в этом случае в числителе получается $2$.

Судя по всему, для функций существует и мнимая область значений.
